```python
import requests
from bs4 import BeautifulSoup

url = "https://udn.com/news/breaknews/1"
html = requests.get(url).text
```

```python
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
sel = "#breaknews > div.context-box__content.story-list__holder.story-list__holder--full > div > div.story-list__text > h2 > a"
target = soup.select(sel)
links = list()
for item in target:
    links.append("https://udn.com" + item['href'])
```

```python
import time
for link in links:
    content = requests.get(link).text
    s = BeautifulSoup(content, 'html.parser')
    content_sel = "body > main > div > section.wrapper-left.main-content__wrapper > section > article > div > section.article-content__editor > p"
    news_content = s.select(content_sel)
    news = ""
    for item in news_content:
        if len(item.text) > 10:
            news = news + item.text
    filename = time.strftime("%Y-%m-%d", time.localtime())+"-"+link.split('?')[0].split('/')[-1]+".txt"
    print(filename + " is writing....")
    with open(filename, "w", encoding='utf-8') as fp:
        fp.write(news)
    time.sleep(3)
print("Done")       
```

```python
import os
data =""
for _, _, filenames in os.walk("UDN_news"):
    for filename in filenames:
        with open("UDN_news\\" + filename, "r", encoding = "utf-8") as fp:
            data = data + fp.read()
print(data)
```
